# Filtering r_dose — renal (CrCl) dose adjustment test cases

Test cases exercising the renal dose adjustment filtering logic used by the DOSAGE validator app. Demonstrates filtering by generic name, CrCl range, disease, route, and weight.

```python
import pandas as pd
import numpy as np
```

```python
# Load renal dose dataset
r_dose = pd.read_csv("r_dose.csv")
```

```python
def filter_r_dose(generic=None, disease=None, crcl=None, route=None, weight=None):
    """
    Filter r_dose by generic, CrCl range, optional disease, route, and weight.
    Returns matching renal-adjusted dosing rows.
    """
    df = r_dose.copy()

    # Filter by generic name
    if generic:
        df = df[df['generic'].str.lower() == generic.lower()]

    # Filter by disease name (optional)
    if disease:
        df = df[df['disease'].str.lower() == disease.lower()]
    
    # Filter by route (optional)
    if route:
        df = df[df['route'].str.upper() == route.upper()]
    
    # Filter by CrCl range (required)
    if crcl is not None:
        df = df[
            df['min_crcl'].astype(float) <= crcl
        ]
        df = df[
            df['max_crcl'].astype(float) >= crcl
        ]
    
    # Optional: filter by weight range if present
    if weight is not None:
        df = df[
            ((df['min_weight'].isna()) | (df['min_weight'].astype(float) <= weight)) &
            ((df['max_weight'].isna()) | (df['max_weight'].astype(float) >= weight))
        ]
    
    return df.reset_index(drop=True)

# ▶ Example usage:
# Filter for Cefepime, CrCl = 25, route IV
result = filter_r_dose(generic="Cefepime", crcl=25, route="IV")

print(result)
```
